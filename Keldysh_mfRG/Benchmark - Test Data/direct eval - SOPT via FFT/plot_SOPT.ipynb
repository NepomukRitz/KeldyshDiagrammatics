```python
file = h5py.File("SIAM_SOPT_FFT.h5") # load h5 file
```

```python
list(file) # look at data
```

```python
v = np.array(file['v']); # load data
PiOE = np.array(file['RePi_a_odd_even']) + 1j*np.array(file['ImPi_a_odd_even'])
PiOO = np.array(file['RePi_a_odd_odd']) + 1j*np.array(file['ImPi_a_odd_odd'])
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(16,4))
fig.suptitle("Bubble from SOPT", fontsize=16)
for i in range(2):
    ax[i].set_xlim(-10,10); 
    ax[i].set_xlabel('$\omega$'); 
ax[0].set_ylabel('Re$\Pi$'); ax[1].set_ylabel('Im$\Pi$');
ax[0].plot(v,real(PiOE),label='OE'); ax[1].plot(v,imag(PiOE),label='OE');
ax[0].plot(v,real(PiOO),label='OO'); ax[1].plot(v,imag(PiOO),label='OO');
ax[0].legend(loc='lower right');
```

```python
data1 = np.loadtxt('Pia_odd_even.dat')
data2 = np.loadtxt('Pia_odd_odd.dat')
data3 = np.loadtxt('Pia_odd_odd_c.dat')
```

```python
w = data1[:,0]

Re_Pia_OE = data1[:,1]
Im_Pia_OE = data1[:,2]

Re_Pia_OO = data2[:,1]
Im_Pia_OO = data2[:,2]

Re_Pia_OO_c = data3[:,1]
Im_Pia_OO_c = data3[:,2]
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(16,4))
fig.suptitle("Bubble from SOPT_Comparison", fontsize=16)
for i in range(2):
    ax[i].set_xlim(-10, 10); 
    ax[i].set_xlabel('$\omega$');
    #ax[i].set_ylim(-0.05, 0);
ax[0].set_ylabel('Re$\Pi$'); ax[1].set_ylabel('Im$\Pi$');
ax[0].plot(v,real(PiOE),label='OE_MATLAB'); ax[1].plot(v,imag(PiOE),label='OE_MATLAB');
ax[0].plot(w,Re_Pia_OE, label='OE_C++');    ax[1].plot(w,Im_Pia_OE, label='OE_C++')
ax[0].plot(v,real(PiOO),label='OO_MATLAB'); ax[1].plot(v,imag(PiOO),label='OO_MATLAB');
ax[0].plot(w,Re_Pia_OO, label='OO_C++');    ax[1].plot(w,Im_Pia_OO, label='OO_C++')
ax[0].plot(w,Re_Pia_OO_c,label='OOC_C++');  ax[1].plot(w,Im_Pia_OO_c,label='OOC_C++');
ax[0].legend(loc='lower right');            ax[1].legend(loc='lower right');
```
